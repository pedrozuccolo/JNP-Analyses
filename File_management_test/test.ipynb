# File management

Files in RADAR are organized as follows:

project_folder/participant-folders/measure-folder/csv-folders

The objetive of this script is to create, for each project, one folder for each measure containing observations from participants

```python
import os
import shutil
```

```python
#iterate over file levels
#android_phone_bluetoothe_devices

measure_name = "android_phone_light"
project_folder = "data_test"
destination = "destination"

#this function retrieves in each participant folder the files for a specified measure and allocates it into another folder
for root, dirs, files in os.walk(project_folder):
    if measure_name in root:
        for participant_file in os.listdir(root):
            old_path = root + "/" + participant_file
            participant = root.split("/")[1]
            new_path = destination + "/" + participant + "_" + measure_name + "_" + participant_file
            print("#############")
            print(f"\n{old_path} \n\n\t\t\t\tmoved to \n\n {new_path}\n\n")
            shutil.move(old_path, new_path)
```
